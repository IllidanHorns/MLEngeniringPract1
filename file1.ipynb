<h2>Boxplot<h2>
<h5>"Boxplot()" - функция для вычесления распределения данных и определения выбросов значений (значений, которые откланяются от общей статистики)<h5>

+ "isin()" - позволяет взять набор значений по опредлённому признаку
+ "median()" - позволяет взять медиану значений относетльно верхнего и нижнего квартилей
+ "idxmax()" - выводит максимальное значение по ключу

```python
plt.figure(figsize=(20, 6))
sns.boxplot(x='Станция метро', y='Минут до метро', data=df[df['Станция метро'].isin(['Кунцевская', 'ВДНХ', 'Деловой центр', 'Ломоносовский проспект', 'Минская', 'Серпуховская', 'Крылатское'])])
```

<h5>График выше отображает связь между станцией метро, возле которой квартира и кол-ом минут до этого метро от квартиры. Соответственно на графике оттображены: значения по верхниму и нужнему квартилям, максимальное и минимальное допустимое значения, мединана и выбросы, несоответсвующие общей статистике. По графику можно увидеть, что наибольший доверительный интервал самый маленький у станции "Деловой центр", но при этом станция имеет множество выбросов, которые по своему значению сильно отличаются от допустимых. Также можно заметить, что на станции метро ВДНХ нет выбросов, а максимальный доверительный интервал не огромный и примерно равный с нижней и верхней стороны. Соответственно можно сделать вывод, что расстояние до метро от квартиры на ВДНХ всегда предсказуемо, что удобно для жизни<h5>

```python
values_medians = df.groupby('Станция метро')['Минут до метро'].median()
```

<h5>Вывод категории с макисмальной медианой<h5>

```python
max_median_category = values_medians.idxmax()
print("Категория с максимальной медианой - " + max_median_category)
```

<h5>Вывод значения максимальной медианы<h5>

```python
max_median_value = values_medians.max()
print(max_median_value)
```

<h5>Нахождение кол-ва выбросов в каждой категории (станция метро)<h5>

```python
def get_outliers(data):
    Q1 = data.quantile(0.25)  
    Q3 = data.quantile(0.75)  
    IQR = Q3 - Q1  
    lower = Q1 - 1.5 * IQR
    upper= Q3 + 1.5 * IQR
    outliers = data[(data < lower) | (data > upper)]
    return outliers

outliners = df.groupby('Станция метро')['Минут до метро'].apply(get_outliers)
```

<h5>Станция метро с максимальным кол-ом выбросов<h5><h5>Максимальное кол-во выбросов<h5>

```python
print(outliners.idxmax())
```

```python
print(outliners.max())
```

<h5>Станция метро с минимальным кол-ом выбросов<h5><h5>Минимальное кол-во выбросов<h5>

```python
print(outliners.idxmin())
```

```python
print(outliners.min())
```

<h2>Heatmap<h2>
<h5>"Heatmap()" - необходима, чтобы строить темпловую карту и находить корреляции между атрибутами значений. Heatmap позволяет в удобной форме представлять уровень зависимости между атрибутами.<h5>

```python
data1 = df.drop(columns=['Тип квартиры','Станция метро','Ремонт', 'Регион'])
sns.heatmap(data1.corr(), annot=True, cmap='coolwarm')
```

<h5>Исходя из результата графика корреляций можно сделать несколько выводов. Наибольшая корреляция встречается между схожими по своему значению признаками. На данном графике видно, что самая большая корреляция заметна между площадью и жилой площадь, а также между ценой на квартиру и площадью, площадью и площадью кухни. Такие результаты очевидны, так как общая площадь квартиры влияет прямым образом на кол-во квадратных метров в жилой части квартиры и кухни. А цена на квартиру практически всегда зависит от площади квартиры.<h5>

<h5>Наименьшая корреляция заметна в тех признаках, которые практически не связаны между собой по логическому смыслу. Одни из таких признаков: кол-во минут до метро и площадь. Данные значения абсолютно не связаны, поэтому создают статистическую погрешность в виде отрицателдьной корреляции.<h5>

<h5>Холодные цвета означают низкую корреляцию между признаками. Соответственно отрицаетельные значения корреляции - отрицательная корреляция. Отрицательная корреляция означает, что при увеличении значения по одному признаку, уменьшается значение по другому.<h5>
